# Classification models for CIFAR-10

## Neural Network Classifier:

```python
def model_nn():
    
    input_layer = Input(shape = x_train.shape[1:], name='Input_MLP')
    
    flatten = Flatten(name='Flattening_MLP')(input_layer)
    
    hidden1 = Dense(256, activation = 'relu', name='Hidden1_NN')(flatten)
    hidden2 = Dense(256, activation = 'relu', name='Hidden2_NN')(hidden1)
    
    output = Dense(num_classes, activation = 'softmax', name='Output_NN')(hidden2)
    
    model = Model(inputs= input_layer, outputs = output)
    
    
    return model
```

## Convolutional Neural Network Classifier:

```python
def model_cnn():
    
    input_layer = Input(shape = x_train.shape[1:], name='Input_MLP') # channels last
    
    conv1 = Conv2D(filters= 16, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv1')(input_layer)
    max_pool1 = MaxPooling2D(pool_size = (2,2), strides = (2,2), padding='valid', name='Pool1')(conv1)

    conv2 = Conv2D(filters = 32, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv2')(max_pool1)
    max_pool2 = MaxPooling2D(pool_size = (2,2), strides = (2,2), padding='valid', name='Pool2')(conv2)

    flattened = Flatten(name='Flatt_CNN')(max_pool2)
    
    fc1 = Dense(256, activation = 'relu', name='FC-1')(flattened)
    
    output = Dense(num_classes, activation = 'softmax', name='Output_CNN')(fc1)
    
    model = Model(inputs= input_layer, outputs = output)
        
    return model
```

```python
from keras import regularizers
def model_cnn_regu():
    
    input_layer = Input(shape = x_train.shape[1:], name='Input_MLP') # channels last
    
    conv1 = Conv2D(filters= 64, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv1',kernel_initializer='he_normal')(input_layer)
    Batch1 = BatchNormalization()(conv1)
    conv2 = Conv2D(filters= 64, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv2',kernel_initializer='he_normal')(Batch1)
    Batch2 = BatchNormalization()(conv2)
    max_pool1 = MaxPooling2D(pool_size = (2,2), strides = 2, padding='valid', name='Pool1')(Batch2)

    conv3 = Conv2D(filters = 128, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv3',kernel_initializer='he_normal')(max_pool1)
    Batch3 = BatchNormalization()(conv3)
    conv4 = Conv2D(filters = 128, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv4',kernel_initializer='he_normal')(Batch3)
    Batch4 = BatchNormalization()(conv4)
    max_pool2 = MaxPooling2D(pool_size = (2,2), strides = 2, padding='valid', name='Pool2')(Batch4)
    
    
    conv5 = Conv2D(filters = 256, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv5',kernel_initializer='he_normal')(max_pool2)
    Batch5 = BatchNormalization()(conv5)
    conv6 = Conv2D(filters = 256, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv6',kernel_initializer='he_normal')(Batch5)
    Batch6 = BatchNormalization()(conv6)
    max_pool3 = MaxPooling2D(pool_size = (2,2), strides = 2, padding='valid', name='Pool3')(Batch6)
    
    flattened = Flatten(name='Flatt_CNN')(max_pool3)
    
    dropout3 = Dropout(0.6)(flattened)
    fc1 = Dense(256, activation = 'relu', name='FC-1')(dropout3)
    
    output = Dense(num_classes, activation = 'softmax', name='Output_CNN')(fc1)
    
    model = Model(inputs= input_layer, outputs = output)
        
    return model
        
    

cnn_model_regu = model_cnn_regu()
config_cnn = 'cnn_re' # give a recognizable name
cnn_logger = TensorBoard(log_dir='logs/cnn_logs/'+config_cnn+'/') 
Adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=1e-5, amsgrad=False)
cnn_model_regu.compile(loss='categorical_crossentropy', metrics = ['accuracy'], optimizer=Adam)
cnn_model_regu.fit(x_train, y_train_categorical, batch_size = 64, epochs = 10, 
          validation_data = (x_test, y_test_categorical), callbacks = [cnn_logger, Acc_Logger], verbose = 1)
```

```python
def model_cnn_aug():

    input_layer = Input(shape = x_train.shape[1:], name='Input_MLP') # channels last
    
    conv1 = Conv2D(filters= 16, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv1')(input_layer)
    max_pool1 = MaxPooling2D(pool_size = (2,2), strides = (2,2), padding='valid', name='Pool1')(conv1)

    conv2 = Conv2D(filters = 32, kernel_size = (3,3), padding='same', activation = 'relu', name='Conv2')(max_pool1)
    max_pool2 = MaxPooling2D(pool_size = (2,2), strides = (2,2), padding='valid', name='Pool2')(conv2)

    flattened = Flatten(name='Flatt_CNN')(max_pool2)
    
    fc1 = Dense(256, activation = 'relu', name='FC-1')(flattened)
    
    output = Dense(num_classes, activation = 'softmax', name='Output_CNN')(fc1)
    
    model = Model(inputs= input_layer, outputs = output)
    
    
    return model
```

## Ramping things up: Autoencoder for Semantic Segmentation


### Semantic Segmentation

Convolutional neural network based models have been built up in wide variety to solve
different tasks. In general there can be named the challenges of Classiffication, Semantic
Segmentation, Object Detection and Instance Segmentation among complex newer ones
like Keypoint Detection or DensePose etc. 
Assigning an object class to be found in a
given image as whole, is designated as classiffication, whereas in semantic segmentation all
pixels have to be identified by the object classes they refer to. In contrast to classiffication
there can be multiple object classes found in one picture.




### Segnet - An Autoencoder for Semantic Segmentation

Based on the [Kitti Road dataset](http://www.cvlibs.net/datasets/kitti/eval_road.php). A segmentation dataset for autonomous driving created by _Karlsruhe Institute of Technology (KIT)_, MPI Tübingen and University of Toronto.


![CNN Autoencoder](img/segnet.png "CNN Autoencoder")


                                    Source: http://mi.eng.cam.ac.uk/projects/segnet/

It is possible to solve this classification task by using a softmax layer in the end or regressing a rgb image given as ground truth. In the latter case, the rgb values might not be exactly the same but the dimension space is way lower and might need some thresholding.
The network architecture of an autoencoder uses some structure that was often trained previously on some data like [ImageNet](http://www.image-net.org/). The idea is that these weights already have something in common with the task later on so that training converges faster and possibly better than starting off with randomized weights. In the SegNet architecture above the standard classification network architecture `VGG-16` is used to encode the input image into higher abstract space. Afterwards Upsampling and Convolutions project extracted features back into the original input space.

## Some final questions:

**Question:** _Explain differences between the regression and classification task?_

**Answer:**  The Ground truth of classification is Bitmap. The Ground truth of regression is RGB. The classification task is to learn the boundary of the objetcs from the different classes. The regression task is to learn the distribution of the color.
The loss function of regression and classification is different.

**Question:** _What would you suggest to improve your segmentation model?_

**Answer:**  We can use larger batch-sizes or more epochs and add learning decay.
